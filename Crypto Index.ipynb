# Creating a crypto Index Fund
On this notebook, we will define a toy cryptocurrency index and simulate the operation of an investment fund based on it. Indices are basically  baskets of asset assembled according to some preset rules. The famous [S&P500](https://en.wikipedia.org/wiki/S%26P_500_Index) index is comprised of 500 american companies ranked by market capitalization.

Our toy index, will be composed of the top 7 cryptocurrencies by market capitalization at the time of this writing. We will then create a hypothetical fund based on it.

## Defining investment strategies
Now that we have our series we can then write some extra code to implement some investment strategies, and then compare their returns. strategies we will compare is Buy and hold (a pre-defined cryptocurrency) and a fictitious index fund which is rebalanced on a regular basis (daily, weekly, etc).
### Strategy 1: Buy and hold
Suppose you had US\$100,000 in the begginning and you buy only one cryptocurrency and hold it until the end.

```python
def buy_hold(coin):
    #coins = 100000/coin.close.iloc[0]
    perc_return = 100*coin.close.iloc[-1]/coin.close.iloc[0]
    return perc_return
```

This are the returns if you bought and held during the entire period in percentage values

```python
for k,c in {'eth':etherD, 'btc':btcD, 'ltc':ltcD, 'xrp':xrpD, 'stellar':stlrD, 'EOS':eosD, 'DASH':dashD, 'XMR':xmrD}.items():
    print ("Return for {}: {:.2f}%".format(k, buy_hold(c)))
```

```python
def rr(w):
    return w[-1]/w[0]

class CryptoIndex:
    """
    crypto index composed of a list of `coins`, ranked by `rankby` and rebalanced every `period` units of time.
    """
    def __init__(self, coins, rankby='volume', period=7, initial_cash = 100000):
        self.coins = coins
        self.prices = [c.close.iloc[0] for c in coins]
        self.rankby = rankby
        self.period = period
        self.initial_cash = initial_cash
        self.value_series = [initial_cash]
        self.weights = np.array([1/len(coins) for c in coins])
        self.weight_series = [self.weights]
        # How many coins we have of each at the start
        self.portfolio = np.array([initial_cash*w/p for w,p in zip(self.weights, self.prices)])
        self._calc_factors(self.rankby)
        
    def _calc_factors(self, factor):
        print("Rebalancing  by {}".format(self.rankby))
        for c in self.coins:
            if factor == 'volume':
                c['factor'] = c.volume.rolling(self.period).sum()
            elif factor == 'return_ratio':
                c['factor'] = c.volume.rolling(self.period).apply(rr, raw=True)
                
    def _rebalance(self,t):
        try:
            self.prices = [c.close.iloc[t] for c in self.coins]
        except IndexError as e:
            print(t, [c.close.iloc[t] for c in self.coins])
            raise(e)
        # Sell all
        current_cash = sum([q*p  for q,p in zip(self.portfolio,self.prices)])
        # Update Weights
        self.weights = np.array([c.factor.iloc[t] for c in self.coins])
        self.weights /= self.weights.sum()
        self.weight_series.append(self.weights)
        # Re-buy
        self.portfolio = np.array([current_cash*w/p for w,p in zip(self.weights, self.prices)])
        
    def _update_rank(self):
        pass
    
    def trade(self):
        for t in range(self.period-1, self.coins[0].shape[0], self.period):
            self._rebalance(t)
            self.current_cash = sum([q*p  for q,p in zip(self.portfolio,self.prices)])
            #print("Current portfolio value: ${:.2f}".format(self.current_cash))
            self.value_series.append(self.current_cash)
        print("Total return: {:.2f}%".format(100*self.current_cash/self.initial_cash))
        print("Portfolio:")
        print(self.portfolio)
            
            
        
```

Let's now run our fund rebalancing it every 7 days based on the average volume in the previous 7 days:

```python
CIF_v = CryptoIndex([etherD,btcD,ltcD,xrpD,stlrD,dashD,xmrD])
CIF_v.trade()
```

We can see that the return beats the majority of buy and hold of a single currency.

```python
%%opts Curve  [height=380 width=1024 tools=['hover'], interpolation='steps-pre', show_grid=True] {+axiswise}
c1 = hv.Curve(CIF_v.value_series)
c1.relabel("Rebalancing by avg volume in the previous 7 days")
```

Now let's run the the fund rebalancing in proportion to the return ratio of the previous 7 days:

```python
CIF_rr = CryptoIndex([etherD,btcD,ltcD,xrpD,stlrD,dashD,xmrD], rankby='return_ratio')
CIF_rr.trade()
```

```python
%%opts Curve  [height=380 width=1024 tools=['hover'], interpolation='steps-pre', show_grid=True] {+axiswise}
c1 = hv.Curve(CIF_rr.value_series)
c1.relabel("Rebalancing by Return Ratio in the previous 7 days")
```
